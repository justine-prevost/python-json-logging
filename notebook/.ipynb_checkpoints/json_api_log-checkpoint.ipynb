```python
import requests
```

```python
try :
    url = requests.get('https://jsonplaceholder.typicode.com/users')
    groot.info("URL works")
except NameError : 
    groot.error("Error from the code itself")
except requests.exceptions.HTTPError: 
    groot.error("The url is wrong")
except requests.exceptions.ConnectionError:
    groot.error("The connection wasn't made")
except requests.exceptions.RequestException: 
    groot.error("There is an error")
```

```python
import json
```

```python
user = url.json()
display(user)
```

```python
import random
u = random.randint(1,100)
print(u)
groot.info("Test random")
```

```python
for u in user : 
    u['age'] = random.randint(18,90)
for u in user :
    display(f"Name : {u['name']}, age : {u['age']}")
```

```python
for u in user :
    address = u.get('address', {})
    address.pop('geo', None)
for u in user : 
    display(u['company'])
```

```python
import jmespath

name_under_fifty = jmespath.search("[? age < `50`].[name,age]", user)
display(name_under_fifty)
```

```python
company_infos = jmespath.search("[].company.[name, catchPhrase]",user)
display(company_infos)
```

```python
company_infos2 = jmespath.search("[].company.[name, catchPhrase].{company_name : [0] , company_cacthPhrase : [1]}", user)
display(company_infos2)
```

```python
have_you_seen_Glenna = [u for u in user if 'Glenna' in u['name']]
print(have_you_seen_Glenna)
```
